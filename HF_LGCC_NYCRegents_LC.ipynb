```python
#passed list of missing values for Python to recognize as standard missing values
df = pd.read_csv ('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv', na_values = missing_values)
```

```python
#total of missing values shows theres a lot of missing data!
df.isnull().sum()
```

```python
#creating subset of data with columns I want to consider
df2=df[['School DBN', 'School Name', 'School Level', 'Regents Exam', 'Year',
       'Total Tested', 'Mean Score']]
```

```python
#oldest year data
df2['Year'].min()
```

```python
#recent year data
df2['Year'].max()
```

```python
#filtering data to keep data from year 2017
df3=df2[df2['Year']==2017]
```

```python
#dropping all missing values from data
df3.dropna()
```

```python
#confirming null values shows we still have some unseen null values in Mean Score
df3.isnull().sum()
```

```python
#cleaning up remaining unseen unexpected missing values in the Mean Score column
df3['Mean Score']=pd.to_numeric(df['Mean Score'], errors='coerce')
```

```python
#creating median value for the Mean Score column
median = df3['Mean Score'].median()
```

```python
#replaced missing values in Mean Score column with median values
df3['Mean Score'].fillna(median, inplace=True)
```

```python
#snapshot of header rows shows Mean Score column has no missing values
df3.head()
```

```python
#school selected for comparison is Baruch College Campus High School
df3.loc[(df['School DBN']=='02M411') & (df['Mean Score'])]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```
